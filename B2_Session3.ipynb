WRITE A CODE FOR PASS OR FAIL, PASS BE 40 OR ABOVE RUN THIS CODE ON A LIST OF STUDENT OF 5 STUDENTS

```python
marks= [35,78,40,22,60]
for i in range(len(marks)):
  if marks[i]>=40:
    print("Student",i+1,"-",marks[i],":Pass")
  else:
    print("Student",i+1,'-',marks[i],":Fail")
```

CODE TO CHECK WHETHER NUMBER IS POSITIVE, NEGATIVE OR ZERO IN A GIVEN LIST OF NUMBER

```python
# List of numbers
numbers = [10, -5, 0, 25, -8]

for num in numbers:
    if num > 0:
        print(num, "- Positive")
    elif num < 0:
        print(num, "- Negative")
    else:
        print(num, "- Zero")
```

**STUDENT NAME AS INPUT**

BASED ON THEIR MARKS YOU HAVE TO CONSIDER A PRINT STATEMENT FOR "PERFORMENCE" MARKS IS MORE THAN 75 THEN PERFORMANCE IS STRONG MARKS IS EQUAL TO OR GREATER THAN 40, PERFORMANCE SHOULD PRINT NEED TO IMPORVEMENT LESS 20 THEN PRINT FAIL ,REPEAT CLASS

ATENDANCE 75- WARNING LOW ATTENDANCE STUDY_HOUR LESS THAN 4 PRINT SHOULD COME WARNING NEEDS MORE STUDY TIME

```python
students = {
    "Aishi": {"study_hours": 5, "attendance": 80, "marks": 85},
    "Diya": {"study_hours": 3, "attendance": 70, "marks": 60},
    "Gauri": {"study_hours": 2, "attendance": 65, "marks": 15},
    "Ananya": {"study_hours": 6, "attendance": 90, "marks": 78},
    "Lakshya": {"study_hours": 4, "attendance": 72, "marks": 45}
}

name = input("Enter student name: ")

if name in students:
    s = students[name]

    print("Marks:", s["marks"])

    if s["marks"] > 75:
        print("Performance: Strong")
    elif s["marks"] >= 40:
        print("Performance: Needs Improvement")
    elif s["marks"] < 20:
        print("Performance: Fail, Repeat Class")

    if s["attendance"] < 75:
        print("Warning: Low Attendance")

    if s["study_hours"] < 4:
        print("Warning: Needs More Study Time")

else:
    print("Student not found")
```

In payment processing systems, certain transactions are flagged for review based on specific patterns.

Write a script that loops through a list of transaction dictionaries. If a transaction amount is greater than 5000 AND the payment method is "Gift Card", append the transaction_id to a new list called flagged_transactions. Once the loop completes, print the flagged transactions list.

```python
transactions = [
    {"transaction_id": "TX101", "amount": 6200, "method": "Credit Card"},
    {"transaction_id": "TX102", "amount": 5500, "method": "Gift Card"},
    {"transaction_id": "TX103", "amount": 1200, "method": "Gift Card"},
    {"transaction_id": "TX104", "amount": 8000, "method": "Gift Card"},
    {"transaction_id": "TX105", "amount": 450, "method": "PayPal"}
]
flagged_transactions = []
for t in transactions:
    if t["amount"] > 5000 and t["method"] == "Gift Card":
        flagged_transactions.append(t["transaction_id"])
print(flagged_transactions)
```
